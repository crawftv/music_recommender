```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
artists_page = 'https://en.wikipedia.org/wiki/List_of_hip_hop_musicians'
```

```python
page = requests.get(artists_page)
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
artists_list = soup.find_all('li')
```

```python
a_artists_list = soup.find_all('div', {"class":"div-col columns column-width"})
```

```python
rappers = []
for a in a_artists_list:
    rappers.append(a.find_all('a'))
```

```python
rappers
```

```python
rap = [ i for item in rappers for i in item ]
```

```python
i = 0
[rap[i]['title'] , rap[i]['href']]
```

```python
rap[81]
```

```python
l = []
for i in range(len(rap)):
    try:
        l1 = [rap[i]['title'], rap[i]['href']]
        l.append(l1)
    except:
        pass
```

```python
len(l)
```

```python
rapper_list = l
```

```python
rapper_list
```

```python
rapper_list = pd.DataFrame(rapper_list, columns=['name','wikipedia_page'])
```

```python
rapper_list["name"]= rapper_list["name"].str.replace("\(rapper\)","")
```

```python
def discog_check(wikipedia_page):
    url = 'https://en.wikipedia.org'+wikipedia_page+"_discography"
    r = requests.get(url)
    
    return r.status_code

def discog_page(wikipedia_page, status_code):
    url = 'https://en.wikipedia.org'+wikipedia_page+"_discography"
    r = requests.get(url)
    
    return r.text
```

```python
discog_check("/wiki/2_Chainz")
```

```python
rapper_list["status_code"] = rapper_list["wikipedia_page"].apply(lambda x: discog_check(x) )
```
